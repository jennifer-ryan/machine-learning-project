```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split 
from tensorflow import keras 
```

```python
data = pd.read_csv("powerproduction.csv")
data[data.power == 0].count()
# There are 49 instances wherer= the power output is zero
```

```python
# Remove these
data = data[data.power != 0]
```

```python
# Splitting data into inputs and outputs
x = data['speed'].values.reshape(-1,1)
y = data['power'].values.reshape(-1,1)

# Splitting data into training and testing variables.
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1)
```

```python
# Define the model
model = keras.models.Sequential()

# Add layers 
model.add(keras.layers.Dense(25, input_dim=1, activation="sigmoid"))
model.add(keras.layers.Dense(1, activation='linear'))

# Compile the model
model.compile(loss="mean_squared_error", optimizer="adam") # Mean squared error loss for regression
```

```python
model.fit(x_train, y_train, epochs=500, batch_size=20)
```

```python
accuracy = model.evaluate(x_train, y_train)
```
